```python
import re
import nltk
```

```python
nltk.download('all')
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
stop_words = set(stopwords.words('english'))
```

```python
stop_words
```

```python
def clean_text(text):
  text=text.lower()
  text=re.sub(r'[^a-z\s]',' ',text)
  tokens=word_tokenize(text)
  token=[i for i in tokens if i not in stop_words]
  return ' '.join(tokens)
```

```python
df['Cleaned_tex']=df['Text'].apply(clean_text)
```

```python
df
```

```python
from textblob import TextBlob
```

```python
def sentiment(text):
 p=TextBlob(text).sentiment.polarity
 if p>0:
  return 'positive'
 elif p<0:
  return 'negative'
 else:
  return 'neutral'
```

```python
df['Sentiment']=df['Cleaned_tex'].apply(sentiment)
```
